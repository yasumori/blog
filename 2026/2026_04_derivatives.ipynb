```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * x

# Generate points for the parabola (expanded range to include x=5)
x_vals = np.linspace(-6, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label='$y = 2x$', color='blue', lw=2, alpha=0.3)

# Get two points of x and corresponding y 
x_1 = 0
x_2 = 2
y_1 = f(x_1)
y_2 = f(x_2)

plt.scatter([x_1, x_2], [y_1, y_2], color='black', zorder=5)
# Draw the "Run" (Horizontal line)
plt.plot([x_1, x_2], [y_1, y_1], color='red', linestyle='--')
# Draw the "Rise" (Vertical line)
plt.plot([x_2, x_2], [y_1, y_2], color='green', linestyle='--')

plt.text((x_1 + x_2)/2, y_1 - 0.5, f'x_2 - x_1 = {x_2 - x_1}', color='red', ha='center')
plt.text(x_2 + 0.1, (y_1 + y_2)/2, f'y_2 - y_1 = {y_2 - y_1}', color='green', va='center')

plt.text(x_1, y_1 - 1, f'P1({x_1}, {y_1})', ha='center')
plt.text(x_2, y_2 + 0.5, f'P2({x_2}, {y_2})', ha='center')

plt.title("Visualising Slope: Rise over Run for $y = 2x$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

points = [2.5, 3.0, 4.0]
names = ["P2", "P3", "P4"]

# Generate points for the parabola (expanded range to include x=5)
x_vals = np.linspace(-6, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label='$y = x^2$', color='blue', lw=2, alpha=0.3)

# Get two points of x and corresponding y 
x_1 = 2
y_1 = f(x_1)

for i, (point, name) in enumerate(zip(points, names)):
    x_2 = point
    y_2 = f(x_2)
    plt.scatter([x_1, x_2], [y_1, y_2], color='black', zorder=5)
    # Draw the "Run" (Horizontal line)
    plt.plot([x_1, x_2], [y_1, y_1], color='red', linestyle='--')
    ## Draw the "Rise" (Vertical line)
    plt.plot([x_2, x_2], [y_1, y_2], color='green', linestyle='--')

    if i == 0:
        plt.text(x_1, y_1 - 1, f'P1({x_1}, {y_1})', ha='center')
    plt.text(x_2, y_2 + 0.5, f'{name}({x_2}, {y_2})', ha='center')

## Formatting the plot
plt.title("$y = x^2$ with P1-P4", fontsize=14)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(-5, 30) # Adjust view to see the points clearly
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x_1 = 2
y_1 = f(x_1)

# h is "closer to 0" but not 0
h_values = [2.0, 1.0, 0.1, 0.0001] 
colors = ['red', 'orange', 'green', 'black']

x_vals = np.linspace(0, 5, 400)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f(x_vals), label='$y = x^2$', color='blue', alpha=0.2)

for h, color in zip(h_values, colors):
    x_2 = x_1 + h
    y_2 = f(x_2)
    
    # Calculate the slope (Rise/Run)
    slope = (y_2 - y_1) / (x_2 - x_1)
    
    # Create a line segment to show this specific slope
    line_x = np.linspace(x_1 - 0.5, x_1 + h + 0.5, 10)
    line_y = slope * (line_x - x_1) + y_1
    
    label = f'h={h:.4f}, Slope={slope:.4f}'
    plt.plot(line_x, line_y, color=color, linestyle='--', label=label)
    plt.scatter([x_2], [y_2], color=color, s=30)

plt.scatter([x_1], [y_1], color='blue', s=100, zorder=5, label='Point of Interest (P1)')
plt.title("As $h \\to 0$, the Slope becomes Instantaneous")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2*x

# The points you requested
test_points = [5, -3, 0]
colors = ['red', 'green', 'purple']

# Generate points for the parabola (expanded range to include x=5)
x_vals = np.linspace(-6, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label='$f(x) = x^2$', color='blue', lw=2, alpha=0.3)

# Loop through each point to plot the tangent line
for x_p, color in zip(test_points, colors):
    y_p = f(x_p)
    slope = df(x_p)
    
    # Define tangent line range (short lines centered at the point)
    t_x = np.linspace(x_p - 1, x_p + 1, 10)
    t_y = slope * (t_x - x_p) + y_p
    
    # Plot the tangent line and the point
    plt.plot(t_x, t_y, color=color, linestyle='--', lw=2, 
             label=f'Tangent at x={x_p} (Slope={slope})')
    plt.scatter([x_p], [y_p], color=color, s=100, zorder=5)

# Formatting the plot
plt.title("Visualising Derivatives: $y' = 2x$ at Specific Points", fontsize=14)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(-5, 30) # Adjust view to see the points clearly
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x

def f_wrong1(x):
    return 2.0 * x

def f_wrong2(x):
    return -2.0 * x

# Generate points for the parabola (expanded range to include x=5)
x_vals = np.linspace(-6, 6, 400)
y_vals_correct = f(x_vals)
y_vals_wrong1 = f_wrong1(x_vals)
y_vals_wrong2 = f_wrong2(x_vals)

data = [
    (1,1), (2,2), (3,3)
]

plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals_correct, label='$y = x$', color='blue', lw=2, alpha=0.3)
plt.plot(x_vals, y_vals_wrong1, label='$y = 2x$', lw=2, alpha=0.3)
plt.plot(x_vals, y_vals_wrong2, label='$y = -2x$', lw=2, alpha=0.3)

# visualise three training examples and errors
for i, (x_point, y_point) in enumerate(data):
    plt.scatter(x_point, y_point, color='black', zorder=5)
    plt.text(x_point+1.10, y_point - 0.25, f'Data Point{i+1}({x_point}, {y_point})', ha='center')

    y_wrong1 = int(f_wrong1(x_point))
    plt.scatter(x_point, y_wrong1, color='blue', zorder=5) 
    plt.text(x_point, y_wrong1 + 1, f'Err({x_point}, {y_wrong1})', ha='center')

    y_wrong2 = int(f_wrong2(x_point))
    plt.scatter(x_point, y_wrong2, color='orange', zorder=5) 
    plt.text(x_point, y_wrong2 - 1, f'Err({x_point}, {y_wrong2})', ha='center')

plt.title("Correct weight $w=1$ and two wrong weights $w=2$ and $w=-2$")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def loss(w):
    coef = 14 / 3
    return coef * ((w-1)**2)

def dl(w):
    # dL/dw
    # then chain rule dL/du * du/dw
    coef = 28 / 3
    return coef * (w-1)

w_vals = np.linspace(-5, 7, 400)
error_vals = loss(w_vals)
data = [
    (1,1), (2,2), (3,3)
]

plt.figure(figsize=(10, 8))
plt.plot(w_vals, error_vals, label='$L(w)$', color='blue', lw=2, alpha=0.3)

wrong_weights = [2, -2]
for i, weight in enumerate(wrong_weights):
    plt.scatter(weight, loss(weight), color='black', zorder=5)
    plt.text(weight+0.75, loss(weight), f'$w = {weight}$', ha='center')

    slope = dl(weight)
    t_w = np.linspace(weight-1, weight+1, 10)    
    t_l = slope * (t_w - weight) + loss(weight)
    plt.plot(t_w, t_l,linestyle='--', lw=2, 
             label=f'Tangent at x={weight} (Slope={slope:.2f})')

more_weights = [4, 1]
for i, weight in enumerate(more_weights):
    plt.scatter(weight, loss(weight), color='blue', zorder=5)
    plt.text(weight, loss(weight)-10, f'$w = {weight}$', ha='center')

    slope = dl(weight)
    t_w = np.linspace(weight-1, weight+1, 10)    
    t_l = slope * (t_w - weight) + loss(weight)
    plt.plot(t_w, t_l,linestyle='--', lw=2, 
             label=f'Tangent at x={weight} (Slope={slope:.2f})')

plt.title("The MSE (loss function) and slopes of $L'(w)$")
plt.xlabel("w")
plt.ylabel("error: $L(w)$")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
```

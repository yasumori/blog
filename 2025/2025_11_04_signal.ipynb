```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple sine wave (representing an analog waveform)
t = np.linspace(0, 1, 1000)  # 1 second of data, sampled at 1000 points
frequency = 5  # Frequency of the sine wave (5 Hz)
amplitude = 1  # Peak amplitude
waveform = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the waveform
plt.figure(figsize=(8, 4))
plt.plot(t, waveform, color='orange')
plt.title("Simple Sine Wave (Looks continuous)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple sine wave (representing an analog waveform)
t = np.linspace(0, 1, 100)  # 1 second of data, sampled at 100 points
frequency = 5  # Frequency of the sine wave (1 Hz)
amplitude = 1  # Peak amplitude
waveform = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the waveform
plt.figure(figsize=(8, 4))
plt.plot(t, waveform, 'o', color='orange', markersize=1)
plt.title("Sine Wave (5 Hz: 100 samples)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# keeping one sample per bottom and top
out = np.argsort(waveform)
top5 = out[:5]
bottom5 = out[-5:]
vals1 = [t[v] for v in top5]
vals2 = [t[v] for v in bottom5]

# add two values at t = 0.0 and t = 1.0
t_new = sorted([0.0] + vals1 + vals2 + [1.0])
t_new = np.array(t_new)

frequency = 5  # Frequency of the sine wave (1 Hz)
amplitude = 1  # Peak amplitude
waveform = amplitude * np.sin(2 * np.pi * frequency * t_new)

# Plot the waveform
plt.figure(figsize=(8, 4))
plt.plot(t_new, waveform, marker='o', color='orange', markersize=5)
plt.title("Minimum Sample Wave (5 Hz: 10 samples)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

```
